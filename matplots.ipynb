```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style

X = np.arange(1,10)
Y = np.arange(11,20)
plt.plot(X,Y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("line plot")
plt.show()

plt.scatter(X,Y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("scatter plot")
plt.show()

style.use('ggplot')
x1=[4, 2, 15]
y2=[13, 6, 23]

x11=[6, 9, 11]
y22=[6, 15, 7]

plt.plot(x1, y2, 'y', label='line one', linewidth=5)
plt.ylabel('Y axis')
plt.xlabel("X axis")
plt.grid(True, color='k')
plt.legend()
plt.show()


plt.plot(x11, y22, 'b', label='line two', linewidth=2)
plt.ylabel('Y axis')
plt.xlabel("X axis")
plt.grid(True, color='g')
plt.legend()
plt.show()


x = np.array([1,2,3,4,7])
y = np.array([5,2,6,24,34])

plt.plot(x,y)
plt.show()

plt.scatter(x,y)


#make a subplot
plt.subplot(2, 2, 1)
plt.plot(x, y, "--r")

plt.subplot(2, 2, 2)
plt.plot(x, y, 'g*--')

plt.subplot(2, 2, 3)
plt.plot(x, y, 'bo')

plt.subplot(2, 2, 4)
plt.plot(x, y, 'g*-')



df01 = pd.read_csv("Iris.csv")
df01.head()

```
