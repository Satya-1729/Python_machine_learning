```python
import numpy as np
import pandas as pd

series1 = pd.Series([13,23,34], index=[3,7,9])
print(series1,"\n\n\n" )

Arrat1 = np.array([100,231,333,555])
series3 =pd.Series(Arrat1,index=["emma","swetha","irirs","see"])
print(series3)
print(series3["swetha"],"\n\n\n")

# dictonary
dict1 = {"satya":34,"ritik":43, "krishna":55}
series4 = pd.Series(dict1)
print(series4,"\n\n\n")

series5 = pd.Series(["NewDelhi","WashingtonDC","London","Paris"], index = ["INDIA","USA","UK","France"])
print(series5)
print(series5[1:3],"\n\n",series5[["USA","UK"]],"\n\n\n")

seriesA = pd.Series([1,2,3,4,5], index=["a","b","c","d","e"])
seriesB = pd.Series([50,20,-10,-50,100], index=["z","y","a","c","e"])
seriesADD= seriesA+seriesB
print(seriesADD,"\n\n\n")
print(seriesA.add(seriesB, fill_value =0))
seriesA = pd.Series([1,2,3,4,5], index=["a","b","c","d","e"])
seriesB = pd.Series([50,20,-10,-50,100], index=["z","y","a","c","e"])
seriessub= seriesA-seriesB
print(seriessub,"\n\n\n")
print(seriesA.sub(seriesB, fill_value =0),"\n\n\n")

array1 =np.array([10,20,30])
array2 =np.array([102,103,124])
array3 =np.array([152,143,144])

DF3= pd.DataFrame(array1)
DF2= pd.DataFrame([array1,array2,array3], columns=["A","B","C"])
DF1= pd.DataFrame([array1,array2,array3])
print(DF3 ,"\n\n\n",DF2,"\n\n\n",DF1),"\n\n\n"

SERIESA = pd.Series([1,2,3,4,5], index=["a","b","c","d","e"])
SERIESB = pd.Series([50,20,-10,-50,100], index=["a","b","c","d","e"])
SERIESC = pd.Series([505,240,-110,-650,1600], index=["z","y","a","c","e"])

DF02 = pd.DataFrame([SERIESA,SERIESB])
DF03 = pd.DataFrame([SERIESA,SERIESB,SERIESC])

print(DF02,"\n",DF03,"\n\n\n")


dict23 = {"Arnab":pd.Series([90,95,97], index=["english","science","hindi"]),
          "Ramit":pd.Series([94,97,55], index=["english","science","hindi"]),
          "Samridhi":pd.Series([99,45,65], index=["english","science","hindi"])}

resultDF = pd.DataFrame(dict23)
print(resultDF)






```
