```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Iris.csv')
df.head()

# Min-Max Normalization

x_min = {}
x_max = {}
x_mean = {}
x_std = {}
print(len(df))
col_name = list(df.columns)
print(col_name,"\n\n\n")

numeric_col = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']

for j in range(len(numeric_col)):
    x_min[numeric_col[j]] = min(df[numeric_col[j]])
    x_max[numeric_col[j]] = max(df[numeric_col[j]])
    x_mean[numeric_col[j]] = df[numeric_col[j]].mean()
    x_std[numeric_col[j]] = df[numeric_col[j]].std()



print("MEAN",x_mean,"\n\n\n")
print("Std deviation", x_std,"\n\n\n")
print("minimum_value", x_min,"\n\n\n")
print("maximum_value", x_max,"\n\n\n")

x_norm_sepal_length = []
x_norm_sepal_width = []
x_norm_petal_length = []
x_norm_petal_width = []

x_z_score_sepal_length = []
x_z_score_sepal_width = []
x_z_score_petal_length = []
x_z_score_petal_width = []

for i in range(len(df)):
  x_norm_sepal_length.append((df['SepalLengthCm'][i]-x_min['SepalLengthCm'])/(x_max['SepalLengthCm']-x_min['SepalLengthCm']))
  x_norm_sepal_width.append((df['SepalWidthCm'][i]-x_min['SepalWidthCm'])/(x_max['SepalWidthCm']-x_min['SepalWidthCm']))
  x_norm_petal_length.append((df['PetalLengthCm'][i]-x_min['PetalLengthCm'])/(x_max['PetalLengthCm']-x_min['PetalLengthCm']))
  x_norm_petal_width.append((df['PetalWidthCm'][i]-x_min['PetalWidthCm'])/(x_max['PetalWidthCm']-x_min['PetalWidthCm']))

  x_z_score_sepal_length.append((df['SepalLengthCm'][i]-x_mean['SepalLengthCm'])/x_std['SepalLengthCm'])
  x_z_score_sepal_width.append((df['SepalWidthCm'][i]-x_mean['SepalWidthCm'])/x_std['SepalWidthCm'])
  x_z_score_petal_length.append((df['PetalLengthCm'][i]-x_mean['PetalLengthCm'])/x_std['PetalLengthCm'])
  x_z_score_petal_width.append((df['PetalWidthCm'][i]-x_mean['PetalWidthCm'])/x_std['PetalWidthCm'])

data_frame = {
    'Id':df['Id'],
    'SepalLengthCm':x_norm_sepal_length,
    'SepalWidthCm':x_norm_sepal_width,
    'PetalLengthCm':x_norm_petal_length,
    'PetalWidthCm':x_norm_petal_width,
    'Species':df['Species']

}

data_frame1 = {
    'Id':df['Id'],
    'SepalLengthCm':x_z_score_sepal_length,
    'SepalWidthCm':x_z_score_sepal_width,
    'PetalLengthCm':x_z_score_petal_length,
    'PetalWidthCm':x_z_score_petal_width,
    'Species':df['Species']


}

df1 = pd.DataFrame(data_frame)
print(df1.head(),"\n\n\n")
print("normalized data frame","\n\n\n")

# df1.to_csv('Iris_norm.csv')

# Standardization or z-score Normalization

df2 = pd.DataFrame(data_frame1)
print(df2.head(),"\n\n\n")


# df1.to_csv('Iris_std.csv')




```
